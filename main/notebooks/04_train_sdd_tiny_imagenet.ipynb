# 04 Train Full SDD on Tiny ImageNet (64x64)

```
import matplotlib.pyplot as plt
from src.experiments.notebook_api import load_cfg, deep_update, run_experiment

cfg = load_cfg('configs/tiny_imagenet.yaml')
cfg = deep_update(cfg, {
    'run_name': 'tiny_imagenet_full_sdd',
    'wandb': {'enabled': True, 'tags': ['tiny-imagenet', 'sdd', 'full']},
})
result = run_experiment(cfg, with_eval=True)
history = result['history']
history.tail()
```

```
ax = history.plot(x='epoch', y=['loss_total', 'loss_mse', 'loss_distill', 'gate_mean'], figsize=(10, 4), title='Tiny ImageNet SDD curves')
ax.grid(True)
plt.show()
```
