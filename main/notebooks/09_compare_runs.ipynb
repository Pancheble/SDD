# 09 Compare Runs

```
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

log_dir = Path('outputs/logs')
frames = []
for csv in sorted(log_dir.glob('*_history.csv')):
    df = pd.read_csv(csv)
    df['run'] = csv.stem.replace('_history', '')
    frames.append(df)

all_hist = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
all_hist.head()
```

```
if not all_hist.empty:
    ax = all_hist.pivot_table(index='epoch', columns='run', values='loss_total').plot(figsize=(10, 4), title='Loss comparison')
    ax.grid(True)
    plt.show()
```
