# 03 Train Full SDD on CIFAR-10

```
import matplotlib.pyplot as plt
from src.experiments.notebook_api import load_cfg, deep_update, run_experiment

cfg = load_cfg('configs/cifar10.yaml')
cfg = deep_update(cfg, {
    'run_name': 'cifar10_full_sdd',
    'wandb': {'enabled': True, 'tags': ['cifar10', 'sdd', 'full']},
})
result = run_experiment(cfg, with_eval=True)
history = result['history']
history.tail()
```

```
ax = history.plot(x='epoch', y=['loss_total', 'loss_mse', 'loss_distill', 'gate_mean'], figsize=(10, 4), title='SDD training curves')
ax.grid(True)
plt.show()
```
