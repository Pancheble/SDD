# 09 Run 비교

`outputs/logs/` 에 저장된 모든 학습 history CSV를 불러와 loss curve를 비교합니다.

```python
from pathlib import Path
import pandas as pd, matplotlib.pyplot as plt

log_dir = Path("outputs/logs")
frames = []
for csv in sorted(log_dir.glob("*_history.csv")):
    df = pd.read_csv(csv)
    df["run"] = csv.stem.replace("_history", "")
    frames.append(df)

if frames:
    all_hist = pd.concat(frames, ignore_index=True)
    print(f"로드된 run: {all_hist['run'].unique().tolist()}")
    print(all_hist.groupby("run")["epoch"].max().rename("max_epoch"))
else:
    all_hist = pd.DataFrame()
    print("outputs/logs/ 에 CSV 파일이 없습니다. 먼저 학습을 실행하세요.")
```

```python
if not all_hist.empty:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    pivot_total = all_hist.pivot_table(index="epoch", columns="run", values="loss_total")
    pivot_total.plot(ax=axes[0], title="Total loss comparison")
    axes[0].grid(True, alpha=0.3); axes[0].set_ylabel("loss")

    pivot_mse = all_hist.pivot_table(index="epoch", columns="run", values="loss_mse")
    pivot_mse.plot(ax=axes[1], title="MSE loss comparison")
    axes[1].grid(True, alpha=0.3); axes[1].set_ylabel("loss")

    plt.tight_layout()
    plt.savefig("outputs/figures/run_comparison.png", dpi=150)
    plt.show()
```
